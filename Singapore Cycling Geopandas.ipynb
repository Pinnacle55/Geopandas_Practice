```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
plt.style.use("ggplot")
plt.rcParams['figure.figsize'] = [15, 8]
```

```python
suburbs = gpd.read_file("sg_shapefiles/singapore_suburb_boundaries.shp")

suburbs.plot()
```

```python
# Get coastline data from: https://www.naturalearthdata.com/downloads/

coastline = gpd.read_file("sg_shapefiles/singapore_coastline.shp")

base = coastline.plot(color = 'dimgrey',
                      edgecolor ='silver',
                      # This is the opacity of the layer
                      alpha = 0.7,
                      linewidth = 0.8)

# Remember to set the appropriate values again
base.set_xlim(103.6, 104.1)
base.set_ylim(1.15, 1.5)
# We can make the background black for easier contrast
base.set_facecolor('black')
# Turn off the gridlines
base.grid(False)
```

```python
# KML files are not normally supported. You need to use fiona to enable a driver that allows you
# to read them
import fiona
gpd.io.file.fiona.drvsupport.supported_drivers['KML'] = 'rw'

cycling = gpd.read_file("sg_shapefiles/SG_cycling_round_island.kml", driver = "KML")
cycling.plot()
```

```python
base = coastline.plot(color = 'dimgrey',
                      edgecolor ='silver',
                      # This is the opacity of the layer
                      alpha = 0.7,
                      linewidth = 0.8)

# Remember to set the appropriate values again
base.set_xlim(103.6, 104.1)
base.set_ylim(1.15, 1.5)
# We can make the background black for easier contrast
base.set_facecolor('black')
# Turn off the gridlines
base.grid(False)
suburbs.plot(ax = base, alpha = 0.5, edgecolor = "black")
cycling.plot(ax = base)
```
